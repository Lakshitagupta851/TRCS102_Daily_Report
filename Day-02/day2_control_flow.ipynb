## 1. Day 1 Recap: Variables, Data Types & Mutability

Let's briefly review the key concepts from Day 1:
- **Variables as References**: In Python, a variable is not a physical "box" holding a value, but rather a reference (nametag) pointing to an object in memory.
- **Data Types**: Python has standard data types: integers (`int`), floating-point numbers (`float`), complex numbers (`complex`), strings (`str`), lists (`list`), tuples (`tuple`), dictionaries (`dict`), sets (`set`), booleans (`bool`), and `NoneType`.
- **Mutability vs. Immutability**:
  - **Immutable** objects (`int`, `float`, `str`, `tuple`, `bool`) cannot be changed in place. Changing them creates a new object in memory.
  - **Mutable** objects (`list`, `dict`, `set`) can be modified in-place without changing their memory address.
- **Type Casting**:
  - **Implicit casting** is done automatically by Python (e.g., `5 + 2.0` results in `7.0`).
  - **Explicit casting** is done manually using constructors like `int()`, `float()`, `str()`, `list()`, `tuple()`, `set()`.
- **PEP 8 Guidelines**: Variable names should be in `snake_case` (lowercase with underscores), constants in `UPPER_SNAKE_CASE`, and classes in `PascalCase`.

Let's run a quick code block to verify these concepts:


````python
num=10
print(type(num))
num=float(num)
print(type(num))
````

````python
# Sharing references (Variables as pointers)
list_a = [1, 2, 3]
list_b = list_a  # points to the same object
list_b.append(4)
print(f"list_a: {list_a} | list_b: {list_b} (Same object? {list_a is list_b})")

# Mutability trap vs. safe copying
list_c = list_a.copy()  # Creates a new list object
list_c.append(5)
print(f"list_a: {list_a} | list_c: {list_c} (Same object? {list_a is list_c})")

# Explicit casting
float_val = 9.8
int_val = int(float_val)  # Truncates decimal part
print(f"Cast {float_val} to int -> {int_val} ({type(int_val)})")

print(~-4)
````

### 2.2 Comparison (Relational) Operators
Used to compare two values, returning a boolean (`True` or `False`):
- Equal to: `==`
- Not equal to: `!=`
- Greater than: `>`
- Less than: `<`
- Greater than or equal to: `≥` (`>=`)
- Less than or equal to: `≤` (`<=`)


````python
# Comparison Operators Example
x = 10
y = 20

print("x==y",x==y)
print("x!=y",x!=y)
print(x>=y)


print(f"x == y (Equal): {x == y}")
print(f"x != y (Not Equal): {x != y}")
print(f"x < y (Less Than): {x < y}")
print(f"x >= y (Greater Than or Equal): {x >= y}")
````

### 2.3 Logical Operators
Used to combine conditional statements:
- `and`: Returns `True` if both statements are true.
- `or`: Returns `True` if at least one statement is true.
- `not`: Inverts the result (turns `True` to `False`, and vice versa).

*Note: Python uses **short-circuit evaluation** for `and`/`or`.*


````python
a=int(input("a"))
b=int(input("b"))
c=int(input("c"))

if(a>b and a>c):
    print("a is greater")
elif(b>a and b>c):
    print("b is greater")
elif(c>a and c>b):
    print("c is greater")
````

````python
# Logical Operators with Short-Circuit Example
is_sunny = True
is_weekend = False

print(is_sunny and is_weekend)
print(is_sunny or is_weekend)
print(not is_sunny)


print(f"is_sunny and is_weekend: {is_sunny and is_weekend}")
print(f"is_sunny or is_weekend: {is_sunny or is_weekend}")
print(f"not is_sunny: {not is_sunny}")
print(~12)
````

### 2.5 Bitwise Operators
Used to perform operations on binary numerals at the bit level:
- Bitwise AND: `&`
- Bitwise OR: `|`
- Bitwise XOR: `^`
- Bitwise NOT: `~`
- Zero-fill left shift: `<<`
- Signed right shift: `>>`


````python
# Bitwise Operators Example
n1 = 5  # binary: 0101
n2 = 3  # binary: 0011
print(f"Bitwise AND (5 & 3): {n1 & n2} (binary: 0001)")
print(f"Bitwise OR (5 | 3): {n1 | n2} (binary: 0111)")
print(f"Bitwise XOR (5 ^ 3): {n1 ^ n2} (binary: 0110)")
print(f"Left Shift (1 << 3): {1 << 3} (binary: 1000)")
````

### 2.6 Membership Operators
Used to test if a sequence is present in an object (like a string, list, tuple, dict, or set):
- `in`: Returns `True` if the value is present in the sequence.
- `not in`: Returns `True` if the value is not present in the sequence.


````python
# Membership Operators Example
fruits = ["apple", "banana", "cherry"]
print("apple" in fruits)
print("kiwi" not in fruits)
print("kiwi" in fruits)



print(f"Is 'banana' in fruits? {'banana' in fruits}")
print(f"Is 'grape' not in fruits? {'grape' not in fruits}")
````

- `is not`: Returns `True` if both variables point to different objects.

## 3. Python Loops: `for` and `while`

Loops allow us to repeat a block of code multiple times. In Python, there are two native loops: `for` and `while`.

### 3.1 The `for` Loop and `range()`
The `for` loop is iterator-based. It traverses a sequence (list, tuple, string, set, dictionary) or an iterator.
We use `range(start, stop, step)` to generate a sequence of numbers:
- `start` (optional): Defaults to 0.
- `stop` (required): Generates numbers up to, but not including, this value.
- `step` (optional): Step value for incrementing (or decrementing).

### 3.2 The `while` Loop
A `while` loop continues executing a block of code as long as a specified condition is `True`. 

**Warning**: You must ensure that the condition is updated inside the loop to eventually evaluate to `False`, otherwise you will create an **infinite loop**.


````python
# Count up to 3
count = 1
while count <= 3:
    print(f"Count: {count}")
    count += 1  # updating loop condition variable
````

### 3.3 How to Use Loops (Practical Iteration & Controls)

#### A. Loop Control Statements:
Python provides control statements to modify the execution flow of loops:


##### 1. The `break` Statement
The `break` statement terminates the loop immediately when triggered, jumping to the code execution directly after the loop.


````python
# Example of break: Exit the loop when number 5 is reached
for num in range(1, 10):
    if num == 5:
        print(f"Found {num}! Exiting the loop early.")
        break
    print(f"Current number: {num}")

print("End of program !!")
````

##### 2. The `continue` Statement
The `continue` statement skips the rest of the current loop iteration and moves directly to the next iteration.


````python
# Example of continue: Skip printing the number 5
for num in range(1, 10):
    if num == 5:
        print(f"Skipping {num}...")
        continue
    print(f"Current number: {num}")
````

##### 3. The `pass` Statement
The `pass` statement is a null operation. It does nothing and acts as a placeholder when the syntax requires a statement, but no action is needed yet.


````python
# Example of pass: Placeholder for future logic inside a block
for num in range(1, 6):
    if num == 3:
        # Placeholder: we will implement special behavior for 3 later
        pass
    else:
        print(f"Current number: {num}")
````

#### B. The Unique Loop `else` Clause:
In Python, loops can have an optional `else` block. The `else` block executes **only if the loop terminates normally** (i.e., completes all iterations without hitting a `break` statement). If a `break` is triggered, the `else` block is skipped.

#### C. Parallel and Indexed Iteration:
- `enumerate(iterable)`: Returns both the index and the value.
- `zip(*iterables)`: Pairs up corresponding items from multiple sequences.

Let's see these control structures in action:


````python
# 1. Loop Else: Searching for a target
fruits = ["apple", "banana", "cherry"]
target = "orange"
for fruit in fruits:
    if fruit == target:
        print(f"Found {target}!")
        break
else:
    # This executes because the loop completed naturally without a 'break'
    print(f"{target} not found in the list.")

# 2. Enumerate: Accessing index and value
for index, fruit in enumerate(fruits):
    print(f"Index {index}: {fruit}")

# 3. Zip: Iterating over multiple lists in parallel
prices = [1.2, 0.5, 2.5]
for fruit, price in zip(fruits, prices):
    print(f"{fruit} costs ${price}")
````

## Question : create a shopping cart 
### press 1 to add a item with price 
### press 2 to show the total of the cart and total items in cart
### press 3 to checkout then ask the customer whether they are gonna pay or cancel
### press 4 to exit (use break )

````python

# ============================================
#        SIMPLE SHOPPING CART PROGRAM
#   Only: while, if-elif-else, variables
# ============================================

total = 0
item_count = 0

print("============================")
print("     SIMPLE SHOPPING CART   ")
print("============================")

while True:
    print("\n--- MENU ---")
    print("1. Add Item")
    print("2. Show Total")
    print("3. Checkout")
    print("4. Exit")

    choice = input("Enter choice (1-4): ")

    # ── Add Item ──────────────────────────────
    if choice == "1":
        item_name  = input("Enter item name  : ")
        item_price = float(input("Enter item price : "))
        item_qty   = int(input("Enter quantity   : "))
        total +=  item_price * item_qty

        item_count += 1

        print(f"'{item_name}' added! ")

    # ── Show Total ────────────────────────────
    elif choice == "2":
        if item_count == 0:
            print("Cart is empty!")
        else:
            print(f"Items in cart : {item_count}")
            print(f"Total amount  : Rs.{total}")

    # ── Checkout ──────────────────────────────
    elif choice == "3":
        if item_count == 0:
            print("Cart is empty! Add items first.")
        else:
            print(f"\nTotal Bill : Rs.{total:.2f}")
            print("1. Pay")
            print("2. Cancel")
            pay_choice = input("Enter choice: ")
            if pay_choice == "1":
                print(f"Payment of Rs.{total:.2f} successful!")
                print("Thank you for shopping!")
                # Reset
                total      = 0
                item_count = 0
            elif pay_choice == "2":
                print("Payment cancelled.")
            else:
                print("Invalid choice.")

    # ── Exit ──────────────────────────────────
    elif choice == "4":
        print("Goodbye!")
        break

    # ── Invalid ───────────────────────────────
    else:
        print("Invalid choice! Enter 1 to 4.")
````

### Solution for Practical Question 2: ATM System Simulator (`while` loop)
Loop balance and options, exiting when Choice = 4.


````python
balance = 1000
mock_actions = [(1, 0), (2, 300), (3, 400), (3, 1200), (4, 0)]
action_index = 0

while True:
    # Display menu and get mock action
    choice, amount = mock_actions[action_index]
    print(f"\n--- ATM Menu (Balance: ${balance}) ---")
    print("1. Check Balance | 2. Deposit | 3. Withdraw | 4. Exit")
    print(f"Selected Choice: {choice} (Amount: ${amount})")
    action_index += 1
    
    # Branch logic
    if choice == 1:
        print(f"Current Balance: ${balance}")
    elif choice == 2:
        balance += amount
        print(f"Successfully deposited ${amount}. New Balance: ${balance}")
    elif choice == 3:
        if amount > balance:
            print(f"Declined: Insufficient funds! Cannot withdraw ${amount}.")
        else:
            balance -= amount
            print(f"Successfully withdrew ${amount}. New Balance: ${balance}")
    elif choice == 4:
        print("Thank you for using the ATM! Exiting transaction.")
        break
    else:
        print("Invalid selection! Please choose options 1-4.")
````
