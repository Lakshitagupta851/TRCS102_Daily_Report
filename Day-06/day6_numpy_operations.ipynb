## 1. Introduction: Why NumPy Arrays?

- **Python Lists**: Store pointers to values scattered across your computer's memory. This makes them flexible but slow for doing heavy calculations.
- **NumPy Arrays**: Store numbers directly side-by-side in a **contiguous block of memory** of a single data type. This allows your CPU to perform operations on the numbers instantly.

Let's import `numpy` (usually as `np`) and create our first arrays:

```python
list1=[10,20,30,40]
arr1=np.array(list1)
print(type(arr1))
```

```python
import numpy as np

# 1. Creating a 1D Array (a simple vector)
arr_1d = np.array([10, 20, 30, 40])
print("1D Array:", arr_1d)
print("Shape (size along dimensions):", arr_1d.shape) # returns (4,)
print("Data Type:", arr_1d.dtype)
print("Number of Dimensions:", arr_1d.ndim)
 
# 2. Creating a 2D Array (a matrix / tabular dataset)
data_matrix = np.array([
    [1.5, 2.0, 3.4],
    [4.0, 5.1, 6.2]
])
print("\n2D Array:\n", data_matrix)
print("Shape (rows, columns):", data_matrix.shape) # returns (2, 3)

# 3. Quick Creators for Initializing Variables
zeros = np.zeros((3, 3))        # 3x3 matrix of zeros
ones = np.ones((2, 4))          # 2x4 matrix of ones
sequence = np.arange(0, 10, 2)  # numbers from 0 to 9 stepping by 2

print("\nZeros Matrix:\n", zeros)
print("ones matrix:")
print(ones)
print("\nSequence:", sequence)
```

## 2. Vectorization: Math Without Loops

If you want to perform math on every item in a Python list, you must write a `for` loop. In NumPy, you can do math operations directly on the array. NumPy applies the operation to every element automatically. This is called **vectorization**.

```python
prices = np.array([100, 200, 300, 400])

# Vectorized operation: apply 10% tax on all prices
taxed_prices = prices * 1.10
print("Original prices:", prices)
print("Taxed prices:   ", taxed_prices)

# Adding two arrays element-wise
a = np.array([1, 2, 3])
b = np.array([10, 20, 30])
print("\nSum of a and b:", a + b)
```

### Speed Comparison: Python Loop vs. NumPy Vectorization
Let's see the speed difference by adding one million numbers together:

```python
import time
import numpy as np
size = 10000000
list1 = list(range(size))
list2 = list(range(size))
arr1 = np.arange(size)
arr2 = np.arange(size)

# Time taken using normal Python loops
start = time.time()
result_list = [list1[i] + list2[i] for i in range(size)]
loop_time = time.time() - start

# Time taken using NumPy vectorized operations
start = time.time()
result_arr = arr1 + arr2
numpy_time = time.time() - start

print(f"Python Loop Time:  {loop_time:.6f} seconds")
print(f"NumPy Vector Time: {numpy_time:.6f} seconds")
print(f"🚀 NumPy is {loop_time / numpy_time:.1f}x faster!")
```

## 3. Indexing & Slicing

In Machine Learning, datasets are 2D grids (matrices):
- **Rows** represent individual data samples (e.g., houses, patients, customers).
- **Columns** represent characteristics or features (e.g., area, number of rooms, price).

We slice arrays using: `array[row_indices, column_indices]`. The colon `:` means "select all" along that direction.

```python
import numpy as np
```

```python
dataset = np.array([
    [25, 50, 710],
    [47, 85, 690],
    [31, 62, 780],
    [19, 22, 620]
])
dataset[0,0]=0
dataset[0,2]=0
dataset[3,0]=0
dataset[3,2]=0
dataset
```

```python
temp=np.array([20,30,10,25,15,28,3])
cold_days=temp[temp<20]
hot_days=temp[temp>25]
print(cold_days)
print(hot_days)
```

```python
# Mock dataset: 4 samples (rows), 3 columns (features)
# Columns: [Age, Yearly Income (in Rs.1000s), Credit Score]
dataset = np.array([
    [25, 50, 710],
    [47, 85, 690],
    [31, 62, 780],
    [19, 22, 620]
])

# 1. Get a single value (Row 1, Column 1 -> Income of second person)
print("Income of second person:", dataset[1, 1])

# 2. Get an entire row (Third person's full details)
print("\nThird person details:", dataset[2, :])

# 3. Get an entire column (All Credit Scores)
print("\nAll credit scores:", dataset[:, 2])

# 4. Splits: Separating Input Features (X) from Target Labels (y)
# Let's say columns 0 and 1 are features (Age, Income), and column 2 is the label (Credit Score).
X = dataset[:, 0:2] # All rows, first two columns (0 and 1)
y = dataset[:, 2]   # All rows, only the last column (index 2)

print("\nFeatures X:\n", X)
print("\nLabels y:", y)
```

## 4. Data Filtering & Cleaning (Boolean Masking & `np.where`)

Datasets often contain mistakes, missing values, or outliers (e.g., negative salaries or zero values where it's impossible). NumPy allows us to filter these out easily.

```python
salaries = np.array([45000, -999, 52000, 61000, -999, 48000])
fixed=np.where(salaries<0,10000,salaries)
print(fixed)


age=np.array([10,-25,67,120,-35,200])
fixed=np.where(age<0,-age,age)
agefinal=np.where(fixed>100,100,fixed)
print(agefinal)
```

```python
import numpy as np
salaries = np.array([45000, -999, 52000, 61000, -999, 48000])

# 1. Find invalid placeholder values (e.g., where salary is -999)
is_invalid = salaries == -999
print("Boolean mask for invalid values:", is_invalid)

# 2. Filter out invalid values (Keep only positive salaries)
salaries[salaries < 0]=20000
print("Cleaned salaries list:", salaries)

# 3. Replace invalid values using np.where(condition, value_if_true, value_if_false)
# Let's replace all negative values with 0, and keep positive salaries as they are
fixed_salaries = np.where(salaries < 0, 0, salaries)
print("Salaries after replacing negatives with 0:", fixed_salaries)
```

```python
salaries=np.array([20000,40000,-20000,50000,-10000])
new=np.where(salaries<0,0,salaries)
print(new)
```

## 5. Statistical Summaries & Standardization (Broadcasting)

NumPy offers functions to calculate dataset metrics. 
- `axis=0`: Perform calculation column-wise (vertical down).
- `axis=1`: Perform calculation row-wise (horizontal across).

```python
import numpy as np
```

```python
x=np.array([10,40,20,30,40])
print(np.mean(x))
print(np.median(x))
```

```python
X = np.array([
    [10.0, 200.0],
    [30.0, 100.0],
    [20.0, 300.0]
])
col_means = np.mean(X, axis=0)
row_means = np.mean(X, axis=1)
# 1. Calculate column-wise means (Average of column 0 and column 1 separately)
print("Column-wise Means:", col_means) # Returns [mean_col0, mean_col1]
print("row wise mean :",row_means)
```

## 6. Reshaping & Flattening

AI models sometimes require inputs of specific shapes (e.g. flattening image pixels from a 2D matrix into a 1D vector).

```python
flat_arr = np.arange(6)
print("Original 1D array:", flat_arr)
# Reshape to a 2D matrix of 2 rows and 3 columns
matrix_2d = flat_arr.reshape(2, 3)
print(matrix_2d)
print("\nReshaped to 2x3 Matrix:\n", matrix_2d)

# Flatten it back to 1D
flattened = matrix_2d.flatten()
print(flattened)
print("\nFlattened array back to 1D:", flattened)
```

## 7. Bonus: Matrix Multiplication (The Dot Product)

In neural networks, we multiply inputs by weight grids to make predictions. In NumPy, matrix multiplication is done using the `@` operator or the `np.dot()` function.

```python
inputs = np.array([1.0, 2.0]) 
print(inputs*5)
```

```python
inputs = np.array([1.0, 2.0]) 
weights = np.array([
    [0.2, 0.8, -0.5],
    [0.5, -0.1, 0.4]
])
print(np.dot(inputs, weights))
# Compute neural output: inputs dot weights
outputs = np.dot(inputs, weights) # or inputs @ weights
print("Neural Layer Outputs:", outputs)
```
