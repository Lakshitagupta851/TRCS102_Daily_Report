## 1. Variables: Pointers to Memory

In many programming languages (like C or Java), a variable is seen as a **labeled box** that holds a value. 
In Python, however, a variable is a **nametag (reference)** attached to an **object** in memory.

When you write:
```python
x = 42
```
Python does three things:
1. Creates an integer object with the value `42` in memory.
2. Creates a variable name `x`.
3. Binds (links) the name `x` to the memory address of the object `42`.

We can inspect the memory address using the `id()` function, and check the type of the object using `type()`.

````python
# Creating variables and inspecting their identities
x = 43
print(x)
print("x")
print(f"x is a variable: {x}")
print(f"Value of x: {x}")
print(f"Data Type of x: {type(x)}")
print("Data Type of x: {type(x)}")
print(f"Memory address of x (id): {id(x)}")
````

### 1.1 Sharing References
When you assign one variable to another, e.g., `y = x`, Python does **not** copy the value. Instead, it makes `y` point to the **same object** that `x` is pointing to.

````python
x = 42
y = x  # y now references the same object as x

print(f"id(x): {id(x)}")
print(f"id(y): {id(y)}")

# Check if they point to the exact same object using the 'is' operator
print(f"Are x and y pointing to the same object? {x is y}")
````

### 1.2 Dynamic Typing
Python is **dynamically typed**, meaning you do not need to declare the variable type beforehand. The type is associated with the *object*, not the *variable name*. A variable name can point to an integer at one point and a string at another.

````python
my_var = 100
print(type(my_var))
my_var = "Now I am a string!"
print(type(my_var))
````

## 3. Mutability vs. Immutability

This is one of the most important concepts to master in Python. 

- **Immutable Objects**: Once created, their value **cannot** be changed in place. 
  - Examples: `int`, `float`, `str`, `tuple`, `bool`, `frozenset`.
- **Mutable Objects**: Their value **can** be changed in place without changing their identity (memory address).
  - Examples: `list`, `dict`, `set`.

Let's see this in action.

### 3.1 Immutable Behavior (e.g., Integer)
Let's modify an integer and observe its memory address before and after.

````python
a = 5
print(f"Initial value of a: {a}, Address: {id(a)}")

a = a + 1
print(f"New value of a: {a}, Address: {id(a)}")
# Notice that the address changed! A new object 6 was created in memory, and 'a' was reassigned.
````

### 3.2 Mutable Behavior (e.g., List)
Let's modify a list and observe its memory address.

````python
my_list = [1, 2, 3]
print(f"Initial list: {my_list}, Address: {id(my_list)}")

my_list.append(4)
print(f"Modified list: {my_list}, Address: {id(my_list)}")
# Notice that the address remained EXACTLY the same! The object was mutated in-place.
````

### 3.3 The Danger of Mutable References
Because mutable objects can be modified in-place, having multiple variables reference the same mutable object can lead to unexpected bugs.

````python
list_a = [10, 20, 30]
list_b = list_a  # Both variables point to the SAME list object
listc=list_a.copy()
listc.append(50)
list_b.append(40)
print(listc)
print(list_b)
print(list_a)
"""
print(f"list_b: {list_b}")
print(f"list_a: {list_a} (Notice list_a was modified as well!)")
print(f"Are list_a and list_b the same object? {list_a is list_b}")"""
````

### 3.4 How to Avoid Mutable Reference Bugs
If you want to copy a mutable object without referencing the same one, you need to create a **copy** of it.

````python
list_a = [10, 20, 30]
list_c = list_a.copy()  # Creates a shallow copy (new object in memory)

list_c.append(40)

print(f"list_c (modified): {list_c}")
print(f"list_a (original): {list_a} (Left untouched!)")
print(f"Are list_a and list_c the same object? {list_a is list_c}")
````

## 4. Sequence & Collection Types

Let's briefly introduce Python's main collection types which are heavily used in AI/ML data preprocessing:

1. **List** (`list`): Ordered, mutable, allows duplicate elements.
2. **Tuple** (`tuple`): Ordered, **immutable**, allows duplicate elements. Think of it as a read-only list.
3. **Dictionary** (`dict`): Unordered (ordered by insertion in Python 3.7+), mutable, key-value pairs, keys must be unique and immutable.
4. **Set** (`set`): Unordered, mutable, unique elements only.

````python
my_list = ["apple", "banana", "cherry", "apple"]
my_tuple = ("apple", "banana", "cherry", "apple")

new=list(my_tuple)
new.append("grapes")
my_tuple=tuple(new)
print(my_tuple)
````

````python
# Defining collections
my_list = ["apple", "banana", "cherry", "apple"]
my_tuple = ("apple", "banana", "cherry", "apple")
my_dict = {"name": "Alice", "age": 25, "role": "Data Scientist"}
my_set = {"apple", "banana", "cherry", "apple"}  # duplicates will be automatically removed
my_list.append(40)
my_tuple.append("grapes") 
print(f"List (duplicates allowed): {my_list}")
print(f"Tuple (immutable list): {my_tuple}")
print(f"Dict (key-value): {my_dict}")
print(f"Set (unique values only): {my_set}")
````

## 5. Type Casting / Conversion

Type casting is the process of converting a variable from one data type to another.

- **Implicit Type Conversion**: Done automatically by Python interpreter (e.g., adding an `int` and a `float` yields a `float`).
- **Explicit Type Conversion**: Done manually using Python constructor functions like `int()`, `float()`, `str()`, `list()`, `tuple()`, `set()`, etc.

````python
# Implicit conversion
num_int = 10
num_float = 5.5
result = num_int + num_float
print(f"Value: {result}, Type: {type(result)} (Implicitly cast to float)")

# Explicit conversion
str_num = "123"
casted_int = int(str_num) 
print(type(casted_int))


print(f"Value: {casted_int}, Type: {type(casted_int)} (Explicitly cast)")

# Converting lists to sets to remove duplicates
numbers_list = [1, 2, 2, 3, 4, 4, 4, 5]
unique_numbers = set(numbers_list)
print(f"Original list: {numbers_list}")
print(f"Unique set: {unique_numbers}")
````

## 6. Variable Naming Rules & Best Practices (PEP 8)

To write clean, readable code, adhere to Python's naming conventions:
1. **Rules (Enforced by Python)**:
   - Must start with a letter or an underscore (`_`).
   - Cannot start with a number.
   - Can only contain alpha-numeric characters and underscores (`A-z`, `0-9`, and `_`).
   - Case-sensitive (`age`, `Age`, and `AGE` are three different variables).
   - Cannot be a reserved keyword (like `if`, `while`, `class`, `import`, `def`).

2. **Best Practices (PEP 8 standard)**:
   - Use **snake_case** for variables and functions (e.g., `learning_rate`, `batch_size`).
   - Use **PascalCase** for classes (e.g., `NeuralNetwork`, `DataPreprocessor`).
   - Use **UPPER_SNAKE_CASE** for constants (e.g., `PI`, `MAX_EPOCHS`).
   - Choose descriptive names (avoid single-letter names like `a`, `b`, `c` unless in loops or quick mathematical operations).

## Solutions to Coding Exercises

Below are the verified solutions and explanations for each of the day's exercises.

### Solution 1: Identify the Identity

- `a is b` outputs `True` because `b = a` assigns `b` to point to the same list object in memory.
- `a == c` outputs `True` because the lists have identical values.
- `a is c` outputs `False` because `c = [1, 2, 3]` instantiates a completely new list object in memory, even though its elements match.

````python
# Verification
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print("a is b:", a is b)
print("a == c:", a == c)
print("a is c:", a is c)
````

### Solution 2: The Immutability Trap

Trying to reassign `fruits[1]` directly will raise a `TypeError` since tuples are immutable.
To solve this, we must cast the tuple to a list (which is mutable), edit the item, and cast it back to a tuple.

````python
fruits = ("apple", "pear", "cherry")

# Convert to list, modify, and convert back
fruits_list = list(fruits)
fruits_list[1] = "banana"
fruits = tuple(fruits_list)

print("Modified tuple:", fruits)
````
